```python
def anagram_checker(word_a, word_b):
    ''' This function checks whether two words are an anagram.
    Words entered must be enclosed in quotations " ".
    Otherwise, function will return nameError'''
    if sorted(word_a) == sorted(word_b):
        return True
    else:
        return False
```

```python
anagram_checker("Listen", "Silent")
```

```python
anagram_checker("listen", "silent")
```

```python
def anagram_checker(word_a, word_b):
    ''' This function checks whether two words are an anagram.
    Words entered must be enclosed in quotations " ".
    Otherwise, function will return nameError'''
    if sorted(str.casefold(word_a)) == sorted(str.casefold(word_b)):
        return True
    else:
        return False
```

```python
anagram_checker("Listen", "Silent")
```
