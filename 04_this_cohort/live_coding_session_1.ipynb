```python
def calc_sales_tax(price, tax_rate=0.13):
    tax_amt = price * tax_rate
    return tax_amt
```

```python
calc_sales_tax(15)
```

```python
calc_sales_tax(5,0.08)
```

```python
def calc_total_bill(price, tax_rate=0.13, tip_rate=0.15):
    tax_amt = price * tax_rate
    tip_amt = price * tip_rate
    return price + tax_amt + tip_amt
```

```python
calc_total_bill(100)
```

```python
calc_total_bill(100,0.2,0.1)
```

```python
100 + 100 * 0.2 + 100 * 0.1
```

```python
calc_total_bill(100, tip_rate=0.18)
```

```python
100 + 100 * 0.13 + 100 * 0.18
```
